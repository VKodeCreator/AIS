```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Загрузка данных Iris
iris = load_iris()
data = iris.data
target = iris.target

# Применение метода k средних
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters)
predicted_labels = kmeans.fit_predict(data)

# Визуализация данных и кластеров
pca = PCA(n_components=2)  # Уменьшение размерности для визуализации до 2D
data_2d = pca.fit_transform(data)

plt.figure(figsize=(12, 5))

# Визуализация реальных кластеров (для сравнения)
plt.subplot(1, 2, 1)
plt.scatter(data_2d[:, 0], data_2d[:, 1], c=target, edgecolor='k', cmap='viridis')
plt.title('True Clusters (for classification)')

# Визуализация предсказанных кластеров
plt.subplot(1, 2, 2)
plt.scatter(data_2d[:, 0], data_2d[:, 1], c=predicted_labels, edgecolor='k', cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title('K-Means Clustering')

plt.show()
```
